```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cycler
from scipy import stats
```

```python
df = pd.read_csv('df1.csv')
```

```python
df
```

```python
# ax will be an array of Axes objects
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_prop_cycle(cycler('color', ['#31686c', '#0066a1', '#ce753b', '#853c42', '#b85696']))

# Call plot() method on the appropriate object
ax.plot(df['year'], df['DP'] * 100, linewidth=0.7, label=r'$DP$')
ax.plot(df['year'], df['DG'] * 100, linewidth=0.7, label=r'$DG$')
ax.plot(df['year'], df['vwRetd'] * 100, linewidth=0.7, label=r'$R$')
ax.plot(df['year'], df['Rf'] * 100, linewidth=0.7, label=r'$R_f$')

ax.set_xlabel('Year', fontsize = 12)
ax.xaxis.set_major_locator(plt.MultipleLocator(10))
ax.yaxis.set_major_locator(plt.MultipleLocator(20))
ax.legend()
ax.set_title(r'Figure: Plot of $DP$, $DG$, and $R$', fontsize = 15);
```

```python
# ax will be an array of Axes objects
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_prop_cycle(cycler('color', ['#31686c', '#0066a1', '#ce753b', '#853c42', '#b85696']))

# Call plot() method on the appropriate object
ax.plot(df['year'], df['DP'] * 100, linewidth=0.7, label=r'$DP$')
ax.plot(df['year'], (df['DG'] - 1) * 100, linewidth=0.7, label=r'$DG - 1$')
ax.plot(df['year'], (df['vwRetd'] - 1) * 100, linewidth=0.7, label=r'$R - 1$')
ax.plot(df['year'], df['rf'] * 100, linewidth=0.7, label=r'$r_f$')

ax.set_xlabel('Year', fontsize = 12)
ax.xaxis.set_major_locator(plt.MultipleLocator(10))
ax.yaxis.set_major_locator(plt.MultipleLocator(20))
ax.legend()
ax.set_title(r'Figure: Plot of $DP$, $DG - 1$, and $R - 1$', fontsize = 15);
```

```python
fig.savefig('q1fig1.pdf', format = 'pdf')
```
