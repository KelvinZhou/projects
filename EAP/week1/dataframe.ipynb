```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vwret = pd.read_csv('vwret.csv')
vwret.rename(columns = {'caldt': 'year'}, inplace = True)
vwret['year'] = vwret['year'].astype(str).str[0:4].astype(int)
```

```python
vwret['vwRetd'] = vwret['vwretd'] + 1
vwret['vwRetx'] = vwret['vwretx'] + 1
```

```python
print(vwret.dtypes, '\n', vwret.head(), '\n', vwret.tail())
```

```python
rf = pd.read_excel('rf.xlsx', sheet_name = 'Annual')[['yyyy', 'Rfree']]
rf.rename(columns = {'yyyy': 'year', 'Rfree': 'rf'}, inplace = True)
rf['rf'] = rf['rf'].shift(1)
rf['Rf'] = rf['rf'] + 1
```

```python
print(rf.dtypes, '\n', rf.head(), '\n', rf.tail())
```

```python
df1 = pd.merge(vwret, rf, on = 'year', how = 'inner')
```

```python
df1
```

```python
df1['DP'] = (df1['vwRetd'] - df1['vwRetx']) / df1['vwRetx']
df1['DG'] = (df1['vwRetd'] - df1['vwRetx']) / (df1['vwRetd'] - df1['vwRetx']).shift(1) * df1['vwRetx'].shift(1)
```

```python
df1
```

```python
df1.to_csv('df1.csv', index = False)
```

```python
cay = pd.read_csv('cay.csv')
cay
```

```python
cay = cay[['date', 'cay=c-0.218 a-0.801y+0.441']]
cay.columns = ['year', 'cay']
cay = cay[cay['year'].str.contains('-12-')]
cay['year'] = cay['year'].str[0:4].astype(int)
cay.reset_index(drop=True, inplace = True)
cay
```

```python
# This is important, because we have return data in 2019, otherwise each regression will miss the last observation
cay.loc[len(cay)]=[2019, np.nan]
cay['year'] = cay['year'].astype(int)
cay
```

```python
df2 = pd.merge(df1, cay, on = 'year', how = 'inner')
df2
```

```python
df2.to_csv('df2.csv', index = False)
```
