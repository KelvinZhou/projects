```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cycler
from scipy import stats
import statsmodels.api as sm
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_prop_cycle(cycler('color', ['#0066a1', '#ce753b', '#b85696', '#31686c', '#853c42']))

for delta in [0.1, 0.01, 0.001]:
    size = int(1 / delta + 1)
    a = np.array([np.linspace(0, 1, size),
                  np.zeros(size)]).transpose()
    a[1:, 1] = stats.norm.rvs(size = size - 1)
    a[:, 1] = a[:, 1] * np.sqrt(delta)
    b = np.cumsum(a[1:, 1] * a[1:, 1])
    ax.plot(a[1:, 0], b, linewidth=0.7, label=r'$\Delta = $' + '{}'.format(delta))

ax.xaxis.set_major_locator(plt.MultipleLocator(0.1))
ax.yaxis.set_major_locator(plt.MultipleLocator(0.1))
ax.legend();
```
