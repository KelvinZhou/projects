```python
import numpy as np
import scipy.optimize
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 5
p = 0.8
h = 1/3
```

```python
def fn(c):
    return (1 - c ** (n/(n-1))) / (n * (1 - c ** (1/(n-1)))) - h
```

```python
c = scipy.optimize.fsolve(fn, 0.05)
c
```

```python
x1 = np.linspace(0, c, num=20)
y1 = x1 ** (1 / (n-1))
```

```python
plt.plot(x1, y1);
```

```python
x2 = np.linspace(c, h, num=20)
y2 = np.full(20, c ** (1 / (n-1)))
plt.plot(x2, y2);
```

```python
df1 = pd.DataFrame({"x": x1, "y": y1})
df2 = pd.DataFrame({"x": x2, "y": y2})
df3 = pd.concat([df1, df2])
```

```python
df3.to_csv('revealdata.csv', index= False)
```
