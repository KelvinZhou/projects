```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vwret = pd.read_csv('vwret.csv')
vwret.rename(columns = {'caldt': 'year'}, inplace = True)
vwret['year'] = vwret['year'].astype(str).str[0:4].astype(int)
```

```python
vwret['vwRetd'] = vwret['vwretd'] + 1
vwret['vwRetx'] = vwret['vwretx'] + 1
```

```python
print(vwret.dtypes, '\n', vwret.head(), '\n', vwret.tail())
```

```python
rf = pd.read_excel('rf.xlsx', sheet_name = 'Annual')[['yyyy', 'Rfree', 'AAA', 'BAA', 'tbl', 'lty']]
rf.rename(columns = {'yyyy': 'year', 'Rfree': 'rf'}, inplace = True)
rf['rf'] = rf['rf'].shift(1)
rf['Rf'] = rf['rf'] + 1
```

```python
rf
```

```python
df1 = pd.merge(vwret, rf, on = 'year', how = 'inner')
```

```python
df1
```

```python
df1['DP'] = (df1['vwRetd'] - df1['vwRetx']) / df1['vwRetx']
df1['DG'] = (df1['vwRetd'] - df1['vwRetx']) / (df1['vwRetd'] - df1['vwRetx']).shift(1) * df1['vwRetx'].shift(1)
```
